```python
plt.figure(figsize=(5,5))

with open(f'wts_t_0.txt', 'r') as f:
    l = f.readlines()
w = np.zeros((3000, 3000))
for i, line in enumerate(l):
    tmp = line.split(',')[:-1]
    for j in range(len(tmp)):
        w[i, j] = float(tmp[j])
plt.imshow(w)
# plt.gca().axes.xaxis.set_visible(False)
# plt.gca().axes.yaxis.set_visible(False)
plt.tight_layout()
```

```python
## II

f, ax = plt.subplots(1,4, figsize=(17,4))
wts = w[:500,:500].flatten()
wts = wts[wts>0]
ax[0].hist(wts, bins=30)
ax[0].set_title('II')

wts = w[500:, 500:].flatten()
wts = wts[wts>0]
ax[1].hist(wts, bins=30)
ax[1].set_title('EE')

wts = w[:500, 500:].flatten()
wts = wts[wts>0]
ax[2].hist(wts, bins=30)
ax[2].set_title('IE')

wts = w[500:, :500].flatten()
wts = wts[wts>0]
ax[3].hist(wts, bins=30)
ax[3].set_title('EI')
```
